Crea un código para mostrar los enlaces de una página.
<a href="">....</a>
cuidado con la diferencia entre  find y find_all
para mostrar un atributo se utilizan [], es decir a['href']

```python
import requests
from bs4 import BeautifulSoup

web='https://valeriemoreau.com/'
pagina=requests.get(web)
#print(pagina.content)
soup=BeautifulSoup(pagina.content,'html.parser')
resultado=soup.find_all('a')
for item in resultado:
    print(item['href'])
#print(resultado['href'])
```

```python
import requests
from bs4 import BeautifulSoup

web='https://valeriemoreau.com/'
pagina=requests.get(web)
soup=BeautifulSoup(pagina.content,'html.parser')
resultado=soup.find_all('img')
for item in resultado:
    print(item['src'])
```

```python
import os
import requests
from datetime import datetime
imagen_internet='https://i.pinimg.com/originals/e2/29/2a/e2292a1ed37546135af18307fe4b7415.png'
ruta='C:\\Users\\Tecnicos\\Desktop\\paula\\programación\\jupyter\\'
carpeta_nueva=ruta+'foto'
if os.path.isdir(carpeta_nueva)!=True:
    os.mkdir(carpeta_nueva)
#print(os.path.isdir(carpeta_nueva))
milisegundos=str(datetime.now().timestamp())
imagen_guardada=carpeta_nueva+'\\pollito_'+milisegundos+'.png'
peticion=requests.get(imagen_internet).content

with open(imagen_guardada, 'wb') as x:
	x.write(peticion)
print('imagen guardada')
```
